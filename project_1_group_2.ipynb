```python
# Import the CSV data for the clean crime data
csv_path_2 = os.path.join("crimes_2011_clean.csv")
crime_df = pd.read_csv(csv_path_2, encoding="utf-8")
crime_df.head()
```

```python
# Add column for Zip Code
crime_df['Zip Code'] = ""
crime_df.head()
```

```python
# Perform Geocode lookup to pull zip code from the Lat/Long data
base_url = "https://maps.googleapis.com/maps/api/geocode/json?"
lat = 0
lng = 0
API_key = "AIzaSyApC6RKbQX-oTKm-pSdntYAAGKXw8se1h8"

from pprint import pprint

query_url = f"{base_url}latlng={lat},{lng}&key={API_key}"
print(query_url)
```

```python
# Run though the rows in the Crimes DataFrame and add the zip code to the Lat/Lng based on the geocode lookup

for index, row in crime_df.iloc[:100].iterrows():
    
    #Define the Latitude and Longitude for the API call
    lat = row['Latitude']
    lng = row['Longitude']
    print(lat, lng)

    # Create query string for API call
    query_url = f"{base_url}latlng={lat},{lng}&key={API_key}"
    print(query_url)

    # Call geocode to get resopnse json based on the inputs of latitude, longitude and API key
    response_json = requests.get(query_url).json()

    #pprint(response_json) - USED JUST TO CHECK RESPONSE
    try:
        zip_code = response_json['results'][0]['address_components'][7]['long_name']
        print(zip_code)
        crime_df.loc[index, 'Zip Code'] = zip_code
    except (KeyError, IndexError):
        print(f"Error on index {index}.")
```
