# Zip Code Lookup by Latitude & Longitude

```python
# Dependencies
import pandas as pd
from uszipcode import SearchEngine
from pprint import pprint
```

```python
# Import the CSV data for the clean crime data
csv_path_2 = os.path.join("crimes_2021_real_clean.csv")
crime_df = pd.read_csv(csv_path_2, encoding="utf-8")
crime_df.head()
```

```python
# Add column for Zip Code
crime_df['Zip Code'] = ""
crime_df.head()
```

```python
# Iteratre through the rows to lookup and append the zip code based on latitude & longitude
sr = SearchEngine()

for index, row in crime_df.iterrows():

    lat = row['Latitude']
    lng = row['Longitude']

    result = sr.by_coordinates(lat, lng, radius=5, returns=1)
    
    try:
        crime_df.loc[index, 'Zip Code'] = result[0].zipcode
    except (KeyError, IndexError):
        print(f"Error on index {index}.")

print('Query for zip code lookup complete')
```
